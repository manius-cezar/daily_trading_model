# Description
This notebook was used to preprocessed data from csv datasets to dataset that will be used by NN models and reference models. Both for Apple stocks prices and BTCUSD cryptocurrency

# Stocks

```python
data_dir = 'data_raw/data_converted_ASK'
historic_data_in_minibatch = 250
ema_lengths = [3, 5, 8, 9, 10, 13, 20, 21, 26, 34, 50, 55, 89, 100, 144, 200, 233, historic_data_in_minibatch]
len(ema_lengths)
```

```python
files = os.listdir(data_dir)
files = [f for f in files if f.endswith('.csv')]

train_files, test_files = train_test_split(files, train_size=0.7, random_state=42)
valid_files, test_files = train_test_split(test_files, train_size=2/3, random_state=42)
len(train_files), len(valid_files), len(test_files)
```

```python
X_train, y_train, features_mean_model_train = segmentate_data_from_files(train_files, data_dir, historic_data_in_minibatch, 1, ema_lengths=ema_lengths, ema_length_for_mean_model=historic_data_in_minibatch)
X_valid, y_valid, features_mean_model_valid = segmentate_data_from_files(valid_files, data_dir, historic_data_in_minibatch, 1, ema_lengths=ema_lengths, ema_length_for_mean_model=historic_data_in_minibatch)
X_test, y_test, features_mean_model_test = segmentate_data_from_files(test_files, data_dir, historic_data_in_minibatch, 1, ema_lengths=ema_lengths, ema_length_for_mean_model=historic_data_in_minibatch)
```

```python
# Save preprocessed data to compressed numpy archives for smaller size
np.savez_compressed('./data/X_train_shuffled_with_mean_model.npz', X_train)
np.savez_compressed('./data/X_valid_shuffled_with_mean_model.npz', X_valid)
np.savez_compressed('./data/X_test_shuffled_with_mean_model.npz', X_test)

np.savez_compressed('./data/y_train_shuffled_with_mean_model.npz', y_train)
np.savez_compressed('./data/y_valid_shuffled_with_mean_model.npz', y_valid)
np.savez_compressed('./data/y_test_shuffled_with_mean_model.npz', y_test)

np.savez_compressed('./data/mean_features_train_shuffled_with_mean_model.npz', features_mean_model_train)
np.savez_compressed('./data/mean_features_valid_shuffled_with_mean_model.npz', features_mean_model_valid)
np.savez_compressed('./data/mean_features_test_shuffled_with_mean_model.npz', features_mean_model_test)
```
